```python
embedding_map = {
    "miniLM": emb_miniLM,
    "bge": emb_bge_small,
    "e5": emb_e5
}
```

```python
faiss_base_path = "/content/drive/MyDrive/faiss_db"
chroma_base_path = "/content/drive/MyDrive/chroma_db"
```

```python
class DBLoader:
    def __init__(self, faiss_path, chroma_path):
        self.faiss_path = faiss_path
        self.chroma_path = chroma_path

    def load(self, name, db_type="faiss"):
        emb_name = name.split("_")[0]
        emb = embedding_map[emb_name]

        if db_type == "faiss":
            return FAISS.load_local(
                f"{self.faiss_path}/{name}",
                emb,
                allow_dangerous_deserialization=True
            )
        else:
            return Chroma(
                persist_directory=f"{self.chroma_path}/{name}",
                embedding_function=emb
            )
```

```python
db_loader = DBLoader(faiss_base_path, chroma_base_path)
```

```python

db = db_loader.load("bge_500", "faiss")
```

```python
class Retriever:
    def __init__(self, db):
        self.db = db

    def get_docs(self, query, k=5):
        return self.db.similarity_search(query, k=k)
```

```python
class PromptBuilder:
    def build(self, query, docs):
        context = "\n\n".join([d.page_content for d in docs])

        return f"""
You are a healthcare assistant.

Answer ONLY using the context below.
If not found, say "I don't know".

Context:
{context}

Question:
{query}

Answer:
"""
```

```python
from groq import Groq
from google.colab import userdata
```

```python
groq_api_key=userdata.get('groq_api_key')
```

```python
client = Groq(api_key=groq_api_key)
```

```python
class LLMEngine:
    def __init__(self, api_key):
        self.client = client

    def generate(self, prompt,model):
        response = self.client.chat.completions.create(
            model=model,
            messages=[
                {"role": "system", "content": "You are a healthcare assistant."},
                {"role": "user", "content": prompt}
            ]
        )
        return response.choices[0].message.content
```

```python
class RAGPipeline:
    def __init__(self, db, api_key):
        self.retriever = Retriever(db)
        self.prompt_builder = PromptBuilder()
        self.llm = LLMEngine(api_key)


    def run(self, query,model):

        docs = self.retriever.get_docs(query)


        prompt = self.prompt_builder.build(query, docs)


        answer = self.llm.generate(prompt,model)

        return answer
```

```python
configs = [
    "miniLM_300", "miniLM_500", "miniLM_1000",
    "bge_300", "bge_500", "bge_1000",
    "e5_300", "e5_500", "e5_1000"
]

query = "What is Influenza?How can we prevent it?"
model = ["openai/gpt-oss-120b","llama-3.3-70b-versatile"]
```

```python
print("*********", model[0],"***************")

for config in configs:
    print("\n======", config, "======")

    db = db_loader.load(config, "faiss")   # or "chroma"
    rag = RAGPipeline(db, groq_api_key)

    answer = rag.run(query,model=model[0])
    print(answer)
    print("Word Length: ",len(answer.split()))
```
